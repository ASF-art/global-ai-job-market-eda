### INTRODUCTION

Artificial Intelligence (AI) has become one of the fastest-growing fields worldwide, creating new career opportunities across industries such as healthcare, finance, technology, education, manufacturing, and consulting. As organizations increasingly adopt AI-driven solutions, the demand for skilled professionals continues to grow, making the AI job market highly competitive and dynamic. Understanding employment trends, salary patterns, workforce characteristics, and job requirements in this evolving market is valuable for job seekers, employers, researchers, and policymakers. This project explores a global AI job market dataset to better understand the characteristics of AI-related employment across different countries and industries.

### OBJECTIVES

- To explore the global AI and data job market by analyzing the distribution of job roles, AI specializations, experience levels, work modes, countries, industries, company sizes, educational requirements, salaries, and employee satisfaction.
- To examine how salary varies across different job roles, experience levels, work modes, company sizes, industries, and educational qualifications.
- To investigate the relationship between employee satisfaction and salary to understand potential associations between compensation and employee well-being.
- To identify correlations among key numerical variables, including salary, experience, company rating, job security, career growth, work-life balance, and employee satisfaction.
- To analyze salary patterns across different job roles and experience levels, as well as across countries and work modes, using multivariate analysis.
- To derive meaningful insights into salary trends, workforce characteristics, and employment patterns in the global AI and data job market.

### DATA OVERVIEW

The Global AI and Data Job Market dataset consists of 90,000 records and 35 features, providing comprehensive information on AI and data-related job opportunities across multiple countries, industries, and organizations. The dataset includes variables related to job characteristics, salary, education, company information, employee satisfaction, and workforce trends, making it suitable for exploratory data analysis of the global AI job market.

Key Features Used in This Analysis
- Job Role – AI and data-related job position.
- AI Specialization – Specialized domain within AI.
- Experience Level – Required experience level for the job.
- Experience Years – Years of professional experience required.
- Salary (USD) – Annual salary offered in US dollars.
- Education Required – Minimum educational qualification required.
- Work Mode – Remote, Hybrid, or Onsite.
- Country – Country where the job is located.
- Industry – Industry sector offering the job.
- Company Size – Organization size.
- Company Rating – Overall company rating.
- Employee Satisfaction – Employee satisfaction score.
- Job Security Score – Measure of perceived job stability.
- Career Growth Score – Indicator of career advancement opportunities.
- Work-Life Balance Score – Measure of work-life balance.



Note: The dataset also contains additional variables, such as Bonus (USD), AI Adoption Score, Hiring Difficulty Score, Automation Risk, Promotion Speed, Cost of Living Index, Offer Acceptance Rate, Economic Index, Vacation Days, Skill Demand Score, and other organizational and economic indicators. These variables were not included in the scope of this exploratory data analysis, as the study primarily focused on employment trends, salary patterns, workforce characteristics, and employee-related factors.

### Univariate Analysis

Univariate Analysis focuses on examining the distribution and characteristics of individual variables independently. This section provides an overview of key features in the dataset, helping to understand their distributions, frequencies, and overall patterns before exploring relationships between variables.

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,
              y='job_role',
              hue='job_role',
              legend=False,
              order=df['job_role'].value_counts().index,
              palette='Set2')
plt.title('Distribution of Job Roles')
plt.xlabel('Count')
plt.ylabel("Job Role")
plt.show()
```

- The dataset contains 8 AI and data-related job roles, including Data Analyst, Data Scientist, AI Engineer, Machine Learning Engineer, NLP Engineer, Computer Vision Engineer, Research Scientist, and Software Engineer (AI).
- The distribution of job roles is almost perfectly balanced, with each role contributing approximately 11,250 records out of the total 90,000 observations.
- This balanced representation minimizes bias toward any particular role and enables fair comparisons in the subsequent analyses.
- Since no single job role dominates the dataset, the insights derived from later analyses (such as salary, experience, and work mode) are likely to be more reliable across different AI and data professions.

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(data=df,
              x='experience_level',
              order=df['experience_level'].value_counts().index,
              palette='Set2')
for container in ax.containers:
    ax.bar_label(container,fontsize=10)

plt.title('Distribution of Experience Level')
plt.xlabel('Experirnce Level')
plt.ylabel('Count')
plt.xticks(rotation=15)
plt.show()
              
```

- Senior-level roles have the highest number of records (22,680), while Entry-level roles have the lowest (22,305).
- The difference between the highest and lowest counts is only 375 jobs, indicating that the dataset is very evenly distributed across all experience levels.
- This balanced distribution makes the dataset suitable for comparing salary, job security, work-life balance, and other job-related metrics across different experience levels without one category dominating the analysis.

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(data=df,
              x='work_mode',
              order=df['work_mode'].value_counts().index,
              palette='Set2')
for container in ax.containers:
    ax.bar_label(container,fontsize=10)

plt.title('Distribution of Work Modes')
plt.xlabel('Work Mode')
plt.ylabel('Count')
plt.show()
```

### Bivariate Analysis

Bivariate Analysis examines the relationship between two variables to identify patterns, comparisons, and potential associations. In this section, salary is compared across different categorical variables, and the relationship between employee satisfaction and salary is explored.

```python
plt.figure(figsize=(12,5))
sns.boxplot(data=df,
              x='job_role',
              y='salary_usd',
              palette='Pastel1')

plt.title('Salary Distribution by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Salary (USD)')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

- Research Scientist roles have the highest median salary among all job roles, indicating they are generally the highest-paying positions in the dataset.
- Data Analyst roles have the lowest median salary, suggesting they are relatively lower-paying compared to other AI and data-related roles.
- Most job roles show a wide salary range, indicating that salaries vary based on factors such as experience, company size, and location.
- Several high-value outliers are present across all job roles, with some salaries approaching $300,000, indicating the availability of premium, high-paying opportunities for specialized professionals.

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,
              x='experience_level',
              y='salary_usd',
              palette='Pastel1')

plt.title('Salary Distribution by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.show()
```

- Lead-level professionals have the highest median salary, indicating that higher experience is associated with better compensation.
- Entry-level employees receive the lowest median salary, reflecting the early stage of their careers.
- Senior-level employees earn more than Mid-level employees but generally less than Lead-level professionals.
- The salary spread increases with experience, suggesting that highly experienced professionals have a wider range of earning opportunities.
- All experience levels contain high-salary outliers, showing that exceptional salaries exist across different career stages, although they become more common at higher experience levels.

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,
              x='work_mode',
              y='salary_usd',
              palette='Pastel1')

plt.title('Salary Distribution by Work Mode')
plt.xlabel('Work Mode')
plt.ylabel('Salary (USD)')
plt.show()
```

- Hybrid roles show a slightly higher median salary, but the difference is not substantial.
- All three work modes have a wide salary range, indicating that compensation is influenced more by factors such as job role, experience level, and company characteristics than by work mode.
- Each work mode contains several high-paying outliers (approaching $300,000), showing that premium salaries are available regardless of whether employees work remotely, on-site, or in a hybrid setup.

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,
              x='company_size',
              y='salary_usd',
              palette='Pastel1')

plt.title('Salary Distribution by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Salary (USD)')
plt.show()
```

### Multivariate Analysis

Multivariate Analysis investigates the combined relationships among multiple variables to uncover deeper patterns that cannot be observed through univariate or bivariate analysis alone. This section uses correlation analysis and aggregated comparisons to better understand how different factors collectively influence salary and workforce characteristics.

```python
selected_cols = ['salary_usd',
    'experience_years',
    'bonus_usd',
    'employee_satisfaction',
    'company_rating',
    'job_security_score',
    'career_growth_score',
    'work_life_balance_score',
    'promotion_speed']

plt.figure(figsize=(8,6))
sns.heatmap(
    df[selected_cols].corr(),
    annot=True,
    cmap='PuBu'
    ,
    fmt='.2f',
    linewidths=0.5)
plt.title('Correlation Between Key Variables')
plt.show()
```

- Salary has a strong positive correlation with bonus (r = 0.75), indicating that employees with higher salaries generally receive higher bonuses.
- Salary is strongly correlated with experience (r = 0.72), suggesting that salary tends to increase as employees gain more years of experience.
- Employee satisfaction shows a moderate positive correlation with salary (r = 0.63), implying that higher-paid employees tend to report greater job satisfaction.
- Experience is moderately correlated with job security (r = 0.60) and bonus (r = 0.55), indicating that experienced employees are more likely to enjoy better job security and higher bonus payouts.
- Job security has a moderate positive relationship with employee satisfaction (r = 0.48), suggesting that employees who feel more secure in their jobs are generally more satisfied.
- Work-life balance has a moderate positive correlation with employee satisfaction (r = 0.43), highlighting its importance in overall employee well-being.
- Company rating, career growth score, and promotion speed exhibit weak or negligible correlations with salary, indicating that these factors alone do not strongly influence employee pay in this dataset.

```python
pivot = df.pivot_table(
    values='salary_usd',
    index='job_role',
    columns='experience_level',
    aggfunc='mean')

plt.figure(figsize=(10,6))
sns.heatmap(
    pivot,
    annot=True,
    cmap='PuBu',
    fmt='.0f')
plt.title('Average Salary by Job Role and Experience Level')
plt.show()
```

- Lead-level professionals earn the highest average salaries across all job roles, with average salaries ranging from approximately USD 102,000 to USD 161,000.
- Research Scientists receive the highest average salary among all roles at the Lead level (USD 161,451), followed by Machine Learning Engineers (USD 150,252) and Computer Vision Engineers (USD 149,703).
- Data Analysts consistently have the lowest average salary across all experience levels, with Entry-level Data Analysts earning approximately USD 45,019.
- Average salary increases with experience across every job role, demonstrating a strong positive relationship between experience level and compensation.
- AI-specialized roles such as Research Scientist, Machine Learning Engineer, Computer Vision Engineer, and NLP Engineer generally receive higher salaries than more general data roles, particularly at the Senior and Lead levels.
- The heatmap indicates that both job role and experience level are major determinants of salary, with salary differences being much larger than those observed across industries, company sizes, or education levels.

```python
pivot = df.pivot_table(
    values='salary_usd',
    index='country',
    columns='work_mode',
    aggfunc='mean')
plt.figure(figsize=(11,6))
sns.heatmap(
    pivot,
    annot=True,
    cmap='PuBu',
    fmt='.0f')
plt.title('Average Salary by Country and Work Mode')
plt.show()
```

- The USA offers the highest average salaries across all work modes, with average salaries exceeding USD 131,000, making it the highest-paying country in the dataset.
- Singapore, Australia, and Canada also provide comparatively high salaries across Hybrid, Onsite, and Remote work arrangements.
- India and Brazil have the lowest average salaries across all work modes, with average salaries below USD 55,000.
- Within each country, the average salaries for Hybrid, Onsite, and Remote work are very similar, suggesting that work mode has only a limited impact on salary compared to the country of employment.
- Salary differences between countries are substantially larger than the differences between work modes within the same country, indicating that geographic location is a stronger determinant of salary than work arrangement.
- Overall, the heatmap suggests that country plays a significant role in compensation, while the choice between Hybrid, Onsite, and Remote work has only a modest effect on average salary.
